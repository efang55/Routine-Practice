### 匯入excel方法二

```python
import pandas as pd
```

```python
rfSheet = pd.read_excel("rf.xlsx")
```

```python
rfSheet.head()
```

```python
rf = pd.DataFrame(rfSheet['一年定存固定利率'])
```

```python
rmSheet = pd.read_excel("rm.xlsx")
```

```python
rmSheet.head()
```

```python
rm = pd.DataFrame(rmSheet['Y9999 加權指數'])
```

```python
betaSheet = pd.read_excel("beta.xlsx")
```

```python
betaSheet.head()
```

```python
beta = betaSheet.iloc[:,1:len(betaSheet.columns)]
```

```python
beta.head()
```

```python
rs = pd.DataFrame(rf.values + beta.values * (rm.values - rf.values), columns=beta.columns)
```

```python
rs.insert(0, 'Date', rfSheet['年/月'])
```
